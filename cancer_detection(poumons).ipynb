# 3-Pre-processing **Data**

```python
from tensorflow.keras.layers import Conv2D,Dense, Dropout,MaxPool2D,Flatten,BatchNormalization
#conv:extraire des caracteristiques des images ,dense:couche de cnx,maxpooling:sous echantillonnage qui réduit les dimensions
#Flatten:couche qui aplatit les données en entrées en un vecteur
#BatchNormalization: couche qui normalise les activations de la couche précédente
```

```python
from tensorflow.keras.models import Sequential #modele en emplimant des couches
from tensorflow.keras.applications import VGG16 #modele de reseau de neurones entrainé
from tensorflow.keras.preprocessing.image import ImageDataGenerator #augmnter les données d'image
from tensorflow.keras.applications.resnet import preprocess_input
# creation de visualisation ,vision par ordinateur
import matplotlib.pyplot as plt
import cv2 as cv
#traitement et evaluation des données
from sklearn.metrics import classification_report,confusion_matrix

from sklearn.utils import shuffle # Import shuffle from sklearn.utils

from keras.utils import to_categorical
```

```python
#def de la forme des entrées de modele
input_shape=(224,224,3)

```

```python
#def du nombre de classe de sortie
num_classes=4
```

```python
#configuration du générateur
trainGenerator= ImageDataGenerator(preprocessing_function=preprocess_input,
                                   rotation_range=10,
                                   width_shift_range=0.3,
                                   height_shift_range=0.3,
                                   shear_range=0.2,
                                   zoom_range=0.1,
                                   horizontal_flip=True,
                                   vertical_flip=True,
                                   dtype='float32',
                                   )
```

```python
valGenrator=ImageDataGenerator(preprocessing_function=preprocess_input,
                             dtype='float32')
```

```python
testGenerator=ImageDataGenerator(preprocessing_function=preprocess_input,
                             dtype='float32')
```

```python
train_data=trainGenerator.flow_from_directory(
    train_path_str,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
val_data=valGenrator.flow_from_directory(
    val_path_str,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
test_data=testGenerator.flow_from_directory(
    test_path_str,
    target_size=(224,224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
VGG16_model=VGG16(
    include_top=False,
    weights='imagenet',
    input_shape=input_shape
)
```

```python
for layer in VGG16_model.layers:
  layer.trainable= False
```

# 4-Création de modéle

```python
#création d'un modéle déquentiel
model=Sequential()

#ajout du modéle VGG16 pré-entrainé sans la couche fully connected
model.add(VGG16_model)

#ajout de la couche de normalisation par lots pour améliorer la vitesse
model.add(BatchNormalization())

#ajout d'une couche de pooling max : reduire la dimension des données
model.add(MaxPool2D(2,2))

#ajout de la couche de flatten : préparer les données à l'entrée pour fully connected
model.add(Flatten())
#premiere couche fully connected avec 1024 neurones
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.3)) #couche de regularisation
#deuxieme couche
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.3))
#troisieme couche
model.add(Dense(128,activation='relu'))
#couche de sortie avec un nombre de neurones
model.add(Dense(num_classes,activation='softmax'))

model.summary()#affichage de modele




```

```python
callbacks=[
    tf.keras.callbacks.EarlyStopping(patience=5),
    tf.keras.callbacks.ModelCheckpoint('model.h5', save_best_only=True),
    tf.keras.callbacks.TensorBoard(log_dir='logs')
]
```

```python
# Compile the model
model.compile(optimizer='adam',  # Choose an optimizer like Adam
              loss='categorical_crossentropy',  # Use categorical crossentropy for multi-class classification
              metrics=['accuracy'])  # Track accuracy during training

# Train the model
resultat = model.fit(train_data, validation_data=val_data, epochs=100, verbose=1)
```

```python
loss, acc = model.evaluate(test_data,verbose = 1)
callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=5),  # Arrête l'entraînement si la métrique surveillée ne s'améliore plus après 5 epochs
    tf.keras.callbacks.ModelCheckpoint('model.h5', save_best_only=True),  # Sauvegarde le modèle dans un fichier 'model.h5'
    tf.keras.callbacks.TensorBoard(log_dir='logs')  # Enregistre les logs pour visualiser les métriques d'entraînement dans TensorBoard
]
loss, acc = model.evaluate(test_data,verbose = 1)
```

```python
import seaborn as sns
predictions_prob = model.predict(test_data)
predictions = np.argmax(predictions_prob , axis = 1)
true_label = test_data.classes
report = classification_report(true_label,predictions)
print(report)
```
